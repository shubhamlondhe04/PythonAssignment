1. Write a program to add two lists index-wise. Create a new list that contains the 0th index item from both the list, then the 1st index item, and so on till the last element. any leftover items will get added at the end of the new list.

	Given List:

	list1 = ["M", "na", "i", "Ra"]
	list2 = ["y", "me", "s", "hul"]

	Output:

	[['M','y'], ['na', me'], ['i', 's'], ['Ra', 'hul']]

```python
list1 = ["M", "na", "i", "Ra"]
list2 = ["y", "me", "s", "hul"]

result = []
length = min(len(list1), len(list2))

for i in range(length):
    result.append([list1[i], list2[i]])
if len(list1) > len(list2):
    result.extend(list1[length:])
elif len(list2) > len(list1):
    result.extend(list2[length:])

print(result)
```

5. You are given a list of integers. You are asked to make a list by running through elements of the list by adding all elements greater and itself.

	i.e. Say given list is [2,4,6,10,1] resultant list will be [22,20,10,23].

	For 1st element 2 ->> these are greater (4+6+10) values and 2 itself so on adding becomes 22.

	For 2nd element 4 ->> greater elements are (6, 10) and 4 itself, so on adding 20

	like wise for all other elememts.

	[2,4,6,10,1]-->[22,20,16,10,23]

```python
list1 = [2, 4, 6, 10, 1]
result = []

for i in list1:
    total = 0
    for j in list1:
        if j >= i:
            total += j
    result.append(total)

print(result)
```

7. Sort a list of alphanumeric strings based on product value of numeric character in it. If in any string there is no numeric character take it's product value as 1.

	Input:

	['1ac21', '23fg', '456', '098d','1','kls']
	Output:

	['456', '23fg', '1ac21', '1', 'kls', '098d']

```python
data = ['1ac21', '23fg', '456', '098d', '1', 'kls']
scored_data = []

for s in data:
    product = 1
    has_digit = False
    for char in s:
        if char.isdigit():
            product = product * int(char)
            has_digit = True

    if not has_digit:
        product = 1

    scored_data.append([product, s])
scored_data.sort(reverse=True)

final_output = []
for pair in scored_data:
    final_output.append(pair[1])

print(final_output)
```

8. Write a program that can find the max number of each row of a matrix
	Example:

	Input:

	[[1,2,3],[4,5,6],[7,8,9]]
	Output:

	[3,6,9]

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
max_list = []
for row in matrix:
    biggest_so_far = row[0]
    for number in row:

        if number > biggest_so_far:
            biggest_so_far = number

    max_list.append(biggest_so_far)

print(max_list)
```

10. Write a program to find set of common elements in three lists using sets.

	Input : ar1 = [1, 5, 10, 20, 40, 80]
			ar2 = [6, 7, 20, 80, 100]
			ar3 = [3, 4, 15, 20, 30, 70, 80, 120]

	Output : [80, 20]

```python

ar1 = [1, 5, 10, 20, 40, 80]
ar2 = [6, 7, 20, 80, 100]
ar3 = [3, 4, 15, 20, 30, 70, 80, 120]

set1 = set(ar1)
set2 = set(ar2)
set3 = set(ar3)

common_elements = set1.intersection(set2, set3)

result = list(common_elements)
print(result)
```

14. Write a Python function that takes a list and returns a new list with unique elements of the first list.
	Exercise 1:

	Input:

	[1,2,3,3,3,3,4,5]
	Output:

	[1, 2, 3, 4, 5]


```python
def get_unique_items(original_list):
    unique_list = []
    for item in original_list:

        if item not in unique_list:
            unique_list.append(item)
    return unique_list

my_numbers = [1, 2, 3, 3, 3, 3, 4, 5]
result = get_unique_items(my_numbers)

print(result)
```

16. Write a Python function to check whether a number is perfect or not.¶
	A Perfect number is a number that is half the sum of all of its positive divisors (including itself).

	Example :

	The first perfect number is 6, because 1, 2, and 3 are its proper positive divisors, and 1 + 2 + 3 = 6.
	Equivalently, the number 6 is equal to half the sum of all its positive divisors: ( 1 + 2 + 3 + 6 ) / 2 = 6.

	The next perfect number is 28 = 1 + 2 + 4 + 7 + 14. This is followed by the perfect numbers 496 and 8128.

```python

def check_if_perfect(num):
    sum_of_divisors = 0
    for i in range(1, num):
        if num % i == 0:

            sum_of_divisors = sum_of_divisors + i
    if sum_of_divisors == num:
        return True
    else:
        return False

number_to_test = 6

if check_if_perfect(number_to_test):
    print(f"{number_to_test} is a Perfect Number!")
else:
    print(f"{number_to_test} is NOT a Perfect Number.")
```

17. Write a Python function to concatenate any no of dictionaries to create a new one.
	Sample Dictionary :
	dic1={1:10, 2:20}
	dic2={3:30, 4:40}
	dic3={5:50,6:60}
	Expected Result : {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}


```python

def combine_all_dicts(*list_of_dicts):
    master_dict = {}

    for current_dict in list_of_dicts:
        master_dict.update(current_dict)
    return master_dict
dic1 = {1: 10, 2: 20}
dic2 = {3: 30, 4: 40}
dic3 = {5: 50, 6: 60}
result = combine_all_dicts(dic1, dic2, dic3)

print(result)
```


18. Write a python function that receives a list of integers and prints out a histogram of bin size 10

	Input:
	[13,42,15,37,22,39,41,50]
	Output:
	{11-20:2,21-30:1,31-40:2,41-50:3}


```python
def create_histogram(numbers):
    histogram = {
        "11-20": 0,
        "21-30": 0,
        "31-40": 0,
        "41-50": 0
    }

    for num in numbers:

        if 11 <= num <= 20:
            histogram["11-20"] = histogram["11-20"] + 1

        elif 21 <= num <= 30:
            histogram["21-30"] = histogram["21-30"] + 1

        elif 31 <= num <= 40:
            histogram["31-40"] = histogram["31-40"] + 1

        elif 41 <= num <= 50:
            histogram["41-50"] = histogram["41-50"] + 1
    print(histogram)

data = [13, 42, 15, 37, 22, 39, 41, 50]
create_histogram(data)
```
